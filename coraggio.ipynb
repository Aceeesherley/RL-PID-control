# <font color='blue'>Ways to Run Python</font>

---
We want to introduce four ways to run Python:

* Interactive Command Line through `python`
* Interactive Command Line through `ipython`
* Scripting
* Jupyter Notebook


You first need to make sure that Python is install on your system.
With Linux/Unix systems, a default Python interpreter comes with the operating system. To test this, from the terminal type:

`which python`

and you will see something like:

`/usr/bin/python`


### <font color='red'> Interactive Command Line through `python` </font>

---

* You can use the interpreter in the interactive mode to test some commands.
* You need to type `python` from the command line.
* You can directly type in Python code, and press `Enter` to get the output.
* You can exit the interactive mode with `quit()` or `exit` command or `^Ctrl + D`.
* The sequence you enter will not be saved if you close the current session.

**Sample Session**

1. Open your terminal/Anaconda prompt and type:

`python`


2. From this interactive shell, you will notice that the prompt is `>>>`. This is now allowing us to type Python code directly and execute it.

3. Now type:

`print('Hello world!')`

4. You should see the output on the screen.

5. To exit the Python shell type `exit()` and press `Return` key.



### <font color='red'>Interactive Command Line through `ipython` </font>

---
* IPython is an interactive shell for the Python programming language that offers enhanced introspection, additional shell syntax, tab completion and rich history.
* It does not come by default with Python.
* IPython gives you all that you get in the basic interpreter but with a lot extra (line numbers, advanced editing, more functions, help functions etc).
* It can be started by typing `ipython` at the command line.
* The main aesthetic difference between the Python interpreter and the enhanced IPython interpreter lies in the command prompt: Python uses `>>>` by default, while IPython uses numbered commands (e.g. `In [1]:`).

**Sample Session**

1. Open your terminal/Anaconda prompt and type:

`ipython`

2. This is an enhanced interactive shell that has many features (tab-completion, woot!). It also has a prompt that is numbered.

3. Now type:

`print('Hello world!')`

4. You should see the output on the screen.

5. To exit the Python shell type `exit()` and press `Return` key.

### <font color='red'> Scripting </font>

---
* Real Python programs are made as scripts and look like simple text files.
* These files are given extensions `.py`.
* You can create text files using whatever text editor you like.
* To run the script you need to use the programming language interpreter and specify the name of the created file as an additional parameter:

```
       %python my_python_file.py
```

**Sample Session**

1. Start a new ASCII/text document named `helloworld.py` and enter the following text:

`print('Hello world!')`

2. Save the document and then in your terminal/Anaconda prompt, go to the directory containing the newlt created file.


3. From the terminal/Anaconda prompt, type:

`python helloworld.py`

4. You should see the output on the screen.

````python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
"""
Simulation Description:
    This script simulates an adaptive PID controller using an Actor-Critic architecture combined with a Radial Basis Function (RBF) network. The goal is to replicate the simulation curves described in the reference paper, illustrating the dynamic tracking behavior, control signals, and adaptive changes in PID parameters (Kp, Ki, Kd) over time.

System Model:
    The system is modeled as a nonlinear second-order differential equation:
        x_dot = v
        v_dot = -1.5*x - 0.8*v + 0.5*sin(x) + u + disturbance
    where:
        x: system position output,
        v: system velocity,
        u: control input,
        disturbance: Gaussian noise representing uncertainties.

Adaptive PID Controller:
    - The controller uses an Actor-Critic architecture with an RBF network.
    - The Actor network outputs adjustments to the PID parameters based on the current state vector (error, error derivative, error integral).
    - The Critic network estimates the value function of the current state.
    - Updates are performed using Temporal Difference (TD) error that is computed from the immediate reward.

Simulation:
    - A step reference signal: 0 before 1 second and 1 after 1 second.
    - Uses Euler integration to update the system states.
    - Additional logging is performed for rewards and TD errors for further analysis.

Run the script to see the simulation curves and analyze the adaptive control behavior.
"""

import numpy as np
import matplotlib.pyplot as plt
import math

class ActorCriticPID:
    def __init__(self, num_rbfs=10, state_dim=3, sigma=1.0,
                 alpha_actor=0.01, alpha_critic=0.005, gamma=0.95):
        """
        Initializes the adaptive PID controller with an Actor-Critic structure.

        Parameters:
            num_rbfs (int): Number of RBF units.
            state_dim (int): Dimension of the state vector (error, error derivative, integral error).
            sigma (float): Standard deviation for the RBF functions.
            alpha_actor (float): Learning rate for the Actor network.
            alpha_critic (float): Learning rate for the Critic network.
            gamma (float): Discount factor.
        """
        self.Kp = 1.0
        self.Ki = 1.0
        self.Kd = 1.0

        self.num_rbfs = num_rbfs
        self.state_dim = state_dim
        self.sigma = sigma
        self.gamma = gamma
        self.alpha_actor = alpha_actor
        self.alpha_critic = alpha_critic

        # Initialize RBF centers uniformly in the range [-2, 2] for each state variable.
        self.centers = np.random.uniform(-2, 2, (num_rbfs, state_dim))
        # Actor weights for adjusting Kp, Ki, Kd (shape: 3 x num_rbfs).
        self.actor_weights = np.zeros((3, num_rbfs))
        # Critic weights for estimating the state value (shape: num_rbfs,).
        self.critic_weights = np.zeros(num_rbfs)

    def compute_phi(self, state):
        """
        Computes the RBF activations for a given state.

        Parameters:
            state (np.array): State vector of shape (state_dim,).

        Returns:
            phi (np.array): Vector of RBF activations (shape: num_rbfs,).
        """
        phi = np.zeros(self.num_rbfs)
        for i, center in enumerate(self.centers):
            diff = state - center
            phi[i] = math.exp(-np.dot(diff, diff) / (2 * self.sigma**2))
        return phi

    def actor_output(self, state):
        """
        Generates the adjustment vector for the PID parameters using the actor network.

        Parameters:
            state (np.array): Current state vector.

        Returns:
            adjustment (np.array): PID parameter adjustments [delta_Kp, delta_Ki, delta_Kd].
        """
        phi = self.compute_phi(state)
        adjustment = np.dot(self.actor_weights, phi)
        return adjustment

    def critic_value(self, state):
        """
        Estimates the state value using the critic network.

        Parameters:
            state (np.array): State vector.

        Returns:
            value (float): Estimated value of the current state.
        """
        phi = self.compute_phi(state)
        value = np.dot(self.critic_weights, phi)
        return value

    def update(self, state, reward, next_state):
        """
        Updates both the actor and critic networks with the Temporal Difference (TD) error.

        Parameters:
            state (np.array): Current state vector [error, error derivative, integral].
            reward (float): Immediate reward after taking action.
            next_state (np.array): Next state vector.

        Returns:
            delta_td (float): The TD error computed from the critic.
        """
        phi = self.compute_phi(state)
        phi_next = self.compute_phi(next_state)
        V = np.dot(self.critic_weights, phi)
        V_next = np.dot(self.critic_weights, phi_next)

        # Compute the TD error.
        delta_td = reward + self.gamma * V_next - V

        # Update Critic weights.
        self.critic_weights += self.alpha_critic * delta_td * phi

        # Update Actor weights.
        for i in range(3):
            self.actor_weights[i, :] += self.alpha_actor * delta_td * phi

        # Adjust PID parameters using actor output.
        adjustment = self.actor_output(state)
        self.Kp += adjustment[0]
        self.Ki += adjustment[1]
        self.Kd += adjustment[2]

        # Clip PID parameters to maintain stability.
        self.Kp = np.clip(self.Kp, 0.1, 10.0)
        self.Ki = np.clip(self.Ki, 0.0, 5.0)
        self.Kd = np.clip(self.Kd, 0.0, 5.0)

        return delta_td  # Return TD error for logging.

def system_dynamics(x, v, u, dt):
    """
    Simulates one time step of the nonlinear system using Euler integration.

    System Dynamics:
        x_dot = v
        v_dot = -1.5*x - 0.8*v + 0.5*sin(x) + u + disturbance

    Parameters:
        x (float): Current system position.
        v (float): Current system velocity.
        u (float): Control input.
        dt (float): Time step.

    Returns:
        (x_new, v_new): Updated state values.
    """
    disturbance = np.random.normal(0, 0.05)  # Gaussian noise disturbance.
    acceleration = -1.5 * x - 0.8 * v + 0.5 * math.sin(x) + u + disturbance
    v_new = v + dt * acceleration
    x_new = x + dt * v_new
    return x_new, v_new

def run_simulation():
    """
    Runs the adaptive PID simulation over the defined time horizon.
    Logs outputs, errors, control actions, rewards, and TD errors.
    """
    total_time = 10.0         # Total simulation time in seconds.
    dt = 0.01                 # Simulation time step.
    steps = int(total_time / dt)

    # Initial system states.
    x = 0.0   # Initial position.
    v = 0.0   # Initial velocity.

    # PID error components.
    previous_error = 0.0
    integral_error = 0.0

    # Data logging lists.
    time_history = []
    position_history = []
    setpoint_history = []
    control_history = []
    error_history = []
    Kp_history = []
    Ki_history = []
    Kd_history = []
    reward_history = []
    td_error_history = []

    # Initialize the adaptive PID controller.
    adaptive_pid = ActorCriticPID(num_rbfs=10, state_dim=3, sigma=1.0,
                                  alpha_actor=0.01, alpha_critic=0.005, gamma=0.95)

    for step in range(steps):
        current_time = step * dt

        # Define the setpoint: step input (0 before 1 second, 1 afterwards).
        setpoint = 0.0 if current_time < 1.0 else 1.0

        # Calculate error and its derivative and integral.
        error = setpoint - x
        error_derivative = (error - previous_error) / dt
        integral_error += error * dt

        # Construct the state vector.
        state = np.array([error, error_derivative, integral_error])

        # Compute control action using conventional PID formula.
        control_input = adaptive_pid.Kp * error + adaptive_pid.Ki * integral_error + adaptive_pid.Kd * error_derivative

        # Simulate the system dynamics for one time step.
        x, v = system_dynamics(x, v, control_input, dt)

        # Compute immediate reward (penalize tracking error and control effort).
        reward = - (error ** 2) - 0.001 * (control_input ** 2)

        # Construct the next state vector.
        next_error = setpoint - x
        next_error_derivative = (next_error - error) / dt
        next_state = np.array([next_error, next_error_derivative, integral_error])

        # Update the adaptive controller and log the TD error.
        td_error = adaptive_pid.update(state, reward, next_state)

        # Log data for this time step.
        time_history.append(current_time)
        position_history.append(x)
        setpoint_history.append(setpoint)
        control_history.append(control_input)
        error_history.append(error)
        Kp_history.append(adaptive_pid.Kp)
        Ki_history.append(adaptive_pid.Ki)
        Kd_history.append(adaptive_pid.Kd)
        reward_history.append(reward)
        td_error_history.append(td_error)

        # Update the previous error.
        previous_error = error

    # After the simulation, plot all the logged data.
    plot_results(time_history, position_history, setpoint_history, control_history,
                 error_history, Kp_history, Ki_history, Kd_history, reward_history, td_error_history)

def plot_results(time_history, position_history, setpoint_history, control_history,
                 error_history, Kp_history, Ki_history, Kd_history, reward_history, td_error_history):
    """
    Plots the simulation results including:
        - System output vs. setpoint.
        - Control signal.
        - Tracking error.
        - Evolution of PID parameters.
        - Reward evolution.
        - TD error evolution.

    Parameters:
        time_history (list): Time stamps.
        position_history (list): System positions.
        setpoint_history (list): Reference setpoints.
        control_history (list): Control inputs.
        error_history (list): Tracking errors.
        Kp_history (list): Proportional gains over time.
        Ki_history (list): Integral gains over time.
        Kd_history (list): Derivative gains over time.
        reward_history (list): Rewards at each time step.
        td_error_history (list): TD errors at each time step.
    """
    plt.figure(figsize=(14, 12))

    # Subplot 1: System Output vs. Setpoint
    plt.subplot(3, 2, 1)
    plt.plot(time_history, position_history, label='System Output', linewidth=2)
    plt.plot(time_history, setpoint_history, '--', label='Setpoint', linewidth=2)
    plt.xlabel('Time (seconds)', fontsize=12)
    plt.ylabel('Output', fontsize=12)
    plt.title('System Output vs. Setpoint', fontsize=14)
    plt.legend()
    plt.grid(True)

    # Subplot 2: Control Signal
    plt.subplot(3, 2, 2)
    plt.plot(time_history, control_history, label='Control Signal', color='red', linewidth=2)
    plt.xlabel('Time (seconds)', fontsize=12)
    plt.ylabel('Control Input', fontsize=12)
    plt.title('Control Signal', fontsize=14)
    plt.legend()
    plt.grid(True)

    # Subplot 3: Tracking Error
    plt.subplot(3, 2, 3)
    plt.plot(time_history, error_history, label='Tracking Error', color='green', linewidth=2)
    plt.xlabel('Time (seconds)', fontsize=12)
    plt.ylabel('Error', fontsize=12)
    plt.title('Tracking Error', fontsize=14)
    plt.legend()
    plt.grid(True)

    # Subplot 4: PID Parameter Evolution
    plt.subplot(3, 2, 4)
    plt.plot(time_history, Kp_history, label='Kp', linewidth=2)
    plt.plot(time_history, Ki_history, label='Ki', linewidth=2)
    plt.plot(time_history, Kd_history, label='Kd', linewidth=2)
    plt.xlabel('Time (seconds)', fontsize=12)
    plt.ylabel('Parameter Value', fontsize=12)
    plt.title('Adaptive PID Parameter Evolution', fontsize=14)
    plt.legend()
    plt.grid(True)

    # Subplot 5: Reward Evolution
    plt.subplot(3, 2, 5)
    plt.plot(time_history, reward_history, label='Reward', color='purple', linewidth=2)
    plt.xlabel('Time (seconds)', fontsize=12)
    plt.ylabel('Reward', fontsize=12)
    plt.title('Reward Evolution', fontsize=14)
    plt.legend()
    plt.grid(True)

    # Subplot 6: TD Error Evolution
    plt.subplot(3, 2, 6)
    plt.plot(time_history, td_error_history, label='TD Error', color='orange', linewidth=2)
    plt.xlabel('Time (seconds)', fontsize=12)
    plt.ylabel('TD Error', fontsize=12)
    plt.title('TD Error Evolution', fontsize=14)
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    run_simulation()
````
